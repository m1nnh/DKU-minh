```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

org_image = cv2.imread(image_path)
b, g, r = cv2.split(org_image)
org_image = cv2.merge((r, g, b))

y, x, layer = org_image.shape
print(y, x, layer)

new_image1 = np.zeros((y, x, layer), np.uint8)
print(new_image1.shape)

start = 0
end = int(x / 4) - 1
divide = int(x / 4)
layer = [_, 0, 2, 1]

for i in range(4):
  
  if i == 0:
    for j in range(3):
      new_image1[:, start : end, j] = org_image[:, start : end, j]
  elif i == 3:
    new_image1[:, start : x, layer[i]] = org_image[:, start : x, layer[i]]
    break
  else:
    new_image1[:, start : end, layer[i]] = org_image[:, start : end, layer[i]]
  
  start += divide
  end += divide

plt.imshow(new_image1)
plt.show()

  

  
        
```
