```python
import cv2
image = cv2.imread('../Dongkeun-OpenCV-ImgData/logo.png')

print('image type = {0}'.format(type(image)))
print('image size = {0}'.format(image.size))
print('image dtype = {0}'.format(image.dtype))
```

```python
(b, g, r) = image[100, 50]

print('(b, g, r) :({0}, {1}, {2})'.format(b, g, r))

blue = image[100, 50, 0]
green = image[100, 50, 1]
red = image[100, 50, 2]

print('blue : {0}, green : {1}, red : {2}'.format(blue, green, red))

```

```python
import matplotlib.pyplot as plt

image[100, 50] = (0, 255, 0)
format_image = image[75 : 125, 25: 75]

plt.imshow(format_image)
plt.show()
```

```python
openCvImage = cv2.imread('../Dongkeun-OpenCV-ImgData/logo.png')
b, g, r = cv2.split(openCvImage)
mergeImage = cv2.merge((r, g, b))

plt.subplot(121)
plt.imshow(openCvImage)
plt.show()
plt.subplot(122)
plt.imshow(mergeImage)
plt.show()

```
